```python
import cv2

HEIGHT=1080
WIDTH=1920

PREFIX='../'
FILES=['butterfly','travel','mario','nature','drone']

OUTPUT_FILE='multipleVideos'
EXTENSION='.mp4'

OUT_VIDEO_FILE_NAME=PREFIX+OUTPUT_FILE+EXTENSION

CAP_SIZE=True
MAX_FRAMES=5000

import numpy as np
fileNames=[]

for f in range(len(FILES)):
    fileNames.append(PREFIX+FILES[f]+EXTENSION)

outputVideo = cv2.VideoWriter(OUT_VIDEO_FILE_NAME,cv2.VideoWriter_fourcc('M','J','P','G'),60,(WIDTH,HEIGHT))


for f in range(len(FILES)):
    inputVideo = cv2.VideoCapture(fileNames[f])
    frameNo=0
    while True:
        ret, frame = inputVideo.read()
        if ret:
            outputVideo.write(frame)
            if CAP_SIZE:
                frameNo+=1
                if frameNo==MAX_FRAMES:
                    break
        else:
            inputVideo.release()
            break
        print("(file,frame) = ",f,frameNo)
outputVideo.close()

print("Finish merging videos")


```
